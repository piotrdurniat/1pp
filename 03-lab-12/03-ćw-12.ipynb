```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# in g
masses = [4.1, 14.0, 31.7]

# in m
ball_1_heights = [0.25, 0.5, 1.0, 1.5, 2.0]
ball_2_heights = [0.5, 1.0, 1.5, 2.0]
ball_3_heights = [1.5, 2.0]

# in cm
ball_1_crater_d = [
    [3.170, 2.880, 2.870, 3.635, 2.965],
    [3.760, 3.845, 3.810, 3.580, 3.665],
    [4.310, 3.910, 4.220, 4.350, 4.060],
    [4.625, 4.510, 4.430, 4.615, 4.550],
    [4.760, 4.875, 4.365, 4.575, 4.880],
]

# in cm
ball_2_crater_d = [
    [4.760, 4.750, 4.920, 4.930, 5.120],
    [5.265, 5.270, 5.380, 5.800, 5.600],
    [6.210, 6.055, 6.355, 6.155, 6.225],
    [6.995, 6.890, 6.885, 6.610, 6.775],
]

# in cm
ball_3_crater_d = [
    [7.310, 7.440, 7.460, 7.375, 7.175],
    [7.980, 7.525, 8.035, 8.040, 7.700],
]

# Convert crater diameters from cm to m
ball_1_crater_d = np.array(ball_1_crater_d) / 100
ball_2_crater_d = np.array(ball_2_crater_d) / 100
ball_3_crater_d = np.array(ball_3_crater_d) / 100

# Convert masses from g to kg
masses = np.array(masses) / 1000

print("Ball 1 crater diameters [m]:")
print(ball_1_crater_d)
print()

print("Ball 2 crater diameters [m]:")
print(ball_2_crater_d)
print()

print("Ball 3 crater diameters [m]:")
print(ball_3_crater_d)
print()

print("Masses [kg]:")
print(masses)
print()


# Calculate average crater diameters for each ball
ball_1_crater_d_avg = np.mean(ball_1_crater_d, axis=1)
ball_2_crater_d_avg = np.mean(ball_2_crater_d, axis=1)
ball_3_crater_d_avg = np.mean(ball_3_crater_d, axis=1)

print("Crater diameters averages:")
print(ball_1_crater_d_avg)
print(ball_2_crater_d_avg)
print(ball_3_crater_d_avg)
print()

ball_1_crater_d_avg_log = [
    float(np.log10(diameter)) for diameter in ball_1_crater_d_avg
]
ball_2_crater_d_avg_log = [
    float(np.log10(diameter)) for diameter in ball_2_crater_d_avg
]
ball_3_crater_d_avg_log = [
    float(np.log10(diameter)) for diameter in ball_3_crater_d_avg
]

print("Crater diameters averages log:")
print(ball_1_crater_d_avg_log)
print(ball_2_crater_d_avg_log)
print(ball_3_crater_d_avg_log)

ball_1_energy = [float(masses[0] * 9.81 * height) for height in ball_1_heights]
ball_2_energy = [float(masses[1] * 9.81 * height) for height in ball_2_heights]
ball_3_energy = [float(masses[2] * 9.81 * height) for height in ball_3_heights]

ball_1_energy_log = [float(np.log10(energy)) for energy in ball_1_energy]
ball_2_energy_log = [float(np.log10(energy)) for energy in ball_2_energy]
ball_3_energy_log = [float(np.log10(energy)) for energy in ball_3_energy]

print("Ball potential energy:")
print(ball_1_energy)
print(ball_2_energy)
print(ball_3_energy)

print("Ball potential energy log:")
print(ball_1_energy_log)
print(ball_2_energy_log)
print(ball_3_energy_log)

ball_1_energy_diameters = list(zip(ball_1_energy, ball_1_crater_d_avg))
ball_2_energy_diameters = list(zip(ball_2_energy, ball_2_crater_d_avg))
ball_3_energy_diameters = list(zip(ball_3_energy, ball_3_crater_d_avg))

all_ball_energy_diameters = [
    *ball_1_energy_diameters,
    *ball_2_energy_diameters,
    *ball_3_energy_diameters,
]


# Extract energies and diameters for plotting
energies = [pair[0] for pair in all_ball_energy_diameters]
diameters = [pair[1] for pair in all_ball_energy_diameters]
```

```python
plt.figure(figsize=(20, 12))
plt.scatter(diameters, energies, c="blue", marker="o")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Åšrednica krateru [m]")
plt.ylabel("Energia [J]")
plt.title("Energia w zaleÅ¼noÅ›ci od Å›rednicy krateru (Skala podwÃ³jnie logarytmiczna)")
plt.grid(True)

log_energies = np.log10(energies)
log_diameters = np.log10(diameters)

slope, intercept = np.polyfit(log_diameters, log_energies, 1)

x_reg = np.linspace(min(log_diameters), max(log_diameters), 100)
y_reg = slope * x_reg + intercept


plt.plot(
    10**x_reg,
    10**y_reg,
    "r-",
    label=f"Prosta regresji: E = {slope:.2f}D + {intercept:.2f} [J]",
)
plt.legend()

print()
print("slope", slope)
print("intercept", intercept)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(diameters, energies, c="blue", marker="o")
plt.xlabel("Åšrednica krateru [m]")
plt.ylabel("Energia [J]")
plt.title("Energia vs Åšrednica krateru (Skala liniowa)")
plt.grid(True)

x_linear = np.linspace(min(diameters), max(diameters), 100)
y_linear = (10**intercept) * x_linear**slope

plt.plot(x_linear, y_linear, "r-", label=f"y = x^{slope:.2f}")
plt.legend()

print("\nFinal energy-diameter relationship:")

print(f"E(d)= 10^{intercept:.4f} * d^{slope:.4f} [J]")
print(f"E(d) = {10**intercept:.4e} * d^{slope:.4f} [J]")
```

## NiepewnoÅ›ci pomiaru


### NiepewnoÅ›Ä‡ standardowa wspÃ³Å‚czynnikÃ³w regresji

$$
s_y = \sqrt{\frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{n-2}}
$$

$$
u_a = s_y \sqrt{\frac{n}{n \sum x_i^2 - \left( \sum x_i \right)^2}}
$$

$$
u_b = s_y \sqrt{\frac{\sum x_i^2}{n \sum x_i^2 - \left( \sum x_i \right)^2}}
$$


```python
y_pred = slope * log_diameters + intercept

n = len(log_diameters)
s_y = np.sqrt(np.sum((log_energies - y_pred) ** 2) / (n - 2))

sum_x = np.sum(log_diameters)
sum_x2 = np.sum(log_diameters**2)

u_a = s_y * np.sqrt(n / (n * sum_x2 - sum_x**2))
u_b = s_y * np.sqrt(sum_x2 / (n * sum_x2 - sum_x**2))

print(f"Standard error of estimate (s_y): {s_y:.4f}")
print(f"Uncertainty of slope (u_a): {u_a:.4f}")
print(f"Uncertainty of intercept (u_b): {u_b:.4f}")
```

## NiepewnoÅ›Ä‡ Å›redniej Å›rednicy kraterÃ³w

$$
u_A(D) = \sqrt{\frac{1}{n-1} \sum_{i=1}^{n} (D_i - \bar{D})^2}
$$

$$
u_B(D) = \frac{\Delta D}{\sqrt{3}}
$$

$$
u_c(D) = \sqrt{u_A(D)^2 + u_B(D)^2}
$$

# NiepewnoÅ›Ä‡ energii potencjalnej

WzÃ³r na energiÄ™ potencjalnÄ…:

$$
E = mgh
$$

Prawo przenoszenia niepewnoÅ›ci:

$$
u_c(E) = \sqrt{\left(\frac{\partial E}{\partial m}\right)^2 u_c(m)^2 + \left(\frac{\partial E}{\partial h}\right)^2 u_c(h)^2}
$$

$$
u_c(E) = \sqrt{g^2h^2 u_c(m)^2 + g^2m^2 u_c(h)^2} \\
= g\sqrt{h^2u_c(m)^2 + m^2u_c(h)^2} = \\
= 9.81 \sqrt{h^2u_c(m)^2 + m^2u_c(h)^2}
$$


```python
u_c_h = 0.06
u_c_m = 0.06 * 10 ** (-3)

g = 9.81
u_c_E_ball_1 = [
    float(g * np.sqrt(h**2 * u_c_m**2 + masses[0] ** 2 * u_c_h**2))
    for h in ball_1_heights
]

u_c_E_ball_2 = [
    float(g * np.sqrt(h**2 * u_c_m**2 + masses[1] ** 2 * u_c_h**2))
    for h in ball_2_heights
]

u_c_E_ball_3 = [
    float(g * np.sqrt(h**2 * u_c_m**2 + masses[2] ** 2 * u_c_h**2))
    for h in ball_3_heights
]

print(u_c_E_ball_1)
print(u_c_E_ball_2)
print(u_c_E_ball_3)


# Example calculation for first ball at first height
h = ball_1_heights[0]
m = masses[0]
example_u_c_E = g * np.sqrt(h**2 * u_c_m**2 + m**2 * u_c_h**2)
print(f"For m = {m*1000:.1f} g and h = {h:.2f} m:")
print(
    f"$u_c(E) = 9.81 \\sqrt{{{h:.2f}^2 \\cdot ({u_c_m:.2e})^2 + ({m:.3f})^2 \\cdot ({u_c_h:.2f})^2}} = {example_u_c_E:.4f}$ J"
)
```
